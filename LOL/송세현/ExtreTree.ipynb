ExtraTree

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
def train_test(df):
  y_train = df['blueWins']
  x_train = df.drop(['blueWins'], axis=1)
  return x_train, y_train
```

CL15

```python
df1 = pd.read_csv("/Users/sehyeon/YSAL/CL15_fix.csv")
df1.dtypes.unique()
X1, y1 = train_test(df1)

x1_train, x1_test, y1_train, y1_test = train_test_split(X1, y1, stratify=y1, test_size = 0.3, random_state=42)

xtree1 = ExtraTreesClassifier(random_state=42)
xtree1.fit(x1_train, y1_train)

print(xtree1.score(x1_train, y1_train))
print(xtree1.score(x1_test, y1_test))

#   
sns.set(style='whitegrid', font_scale=0.8)
plt.figure(figsize=(16, 10))
df_importances1 = pd.DataFrame({'feature': X1.columns, 'importance': xtree1.feature_importances_})
df_importances1.sort_values('importance', ascending=False, inplace=True)
ax1 = sns.barplot(x='importance', y='feature', data=df_importances1)
ax1.set_xlabel('Importance')
ax1.set_ylabel('Feature')
plt.show()
```
